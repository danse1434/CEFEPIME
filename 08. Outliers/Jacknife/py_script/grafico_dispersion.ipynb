```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages # Para varios
import numpy as np
```

```python
df1 = pd.read_csv("../Figura/pca2.csv")
```

```python
fig = plt.figure(figsize = [10,8])

# fig = plt.figure()
gpc1 = fig.add_subplot(2,2,1)
gpc1.vlines([0], 0, 1, transform=gpc1.get_xaxis_transform(), colors='black', ls='--')
gpc1.hlines([0], 0, 1, transform=gpc1.get_yaxis_transform(), colors='black', ls='--')
gpc1.set(xlim = [-6,+6], ylim = [-6,+6], xlabel='PC1', ylabel='PC2')
gpc1.scatter(df1.iloc[:,[0,1,2]]['PC1'], df1.iloc[:,[0,1,2]]['PC2'], marker='.')
gpc1.plot(0.407913053, 0.027167572, 'go', color='r', marker='*')

gpc2 = fig.add_subplot(2,2,2)
gpc2.set(xlim = [-6,+6], ylim = [-6,+6], xlabel='PC1', ylabel='PC3')
gpc2.vlines([0], 0, 1, transform=gpc2.get_xaxis_transform(), colors='black', ls='--')
gpc2.hlines([0], 0, 1, transform=gpc2.get_yaxis_transform(), colors='black', ls='--')
gpc2.scatter(df1.iloc[:,[0,1,2]]['PC1'], df1.iloc[:,[0,1,2]]['PC3'], marker='.')
gpc2.plot(0.407913053, -0.054742583, 'go', color='r', marker='*')

gpc3 = fig.add_subplot(2,2,3)
gpc3.set(xlim = [-6,+6], ylim = [-6,+6], xlabel='PC2', ylabel='PC3')
gpc3.vlines([0], 0, 1, transform=gpc3.get_xaxis_transform(), colors='black', ls='--')
gpc3.hlines([0], 0, 1, transform=gpc3.get_yaxis_transform(), colors='black', ls='--')
gpc3.scatter(df1.iloc[:,[0,1,2]]['PC2'], df1.iloc[:,[0,1,2]]['PC3'], marker='.')
gpc3.plot(0.027167572, -0.054742583, 'go', color='r', marker='*')

# Plano de ejes
(xx,yy) = np.meshgrid(np.linspace(-6, 6, 10),np.linspace(-6, 6, 10))
zz = yy * 0


xs = np.array((df1.PC1))
ys = np.array((df1.PC2))
zs = np.array((df1.PC3))
zdir = np.full((15), None)
ls = np.linspace(1,14,14, dtype=np.int32)

x = np.linspace(-6, +6, 10)

ax = fig.add_subplot(2,2,4, projection='3d')
ax.plot_surface(zz,yy,xx,alpha=0.1, color = 'blue')
ax.plot_surface(xx,zz,yy,alpha=0.1, color = 'blue')
ax.plot_surface(xx,yy,zz,alpha=0.1, color = 'blue')


ax.plot(x, np.zeros(x.size), np.zeros(x.size), color = 'blue', ls='--',alpha=0.1)
ax.plot(np.zeros(x.size), x, np.zeros(x.size), color = 'blue', ls='--',alpha=0.1)
ax.plot(np.zeros(x.size), np.zeros(x.size), x, color = 'blue', ls='--',alpha=0.1)


ax.scatter(df1.iloc[:,[0,1]]['PC1'], df1.iloc[:,[0,1]]['PC2'], np.full((15),-6),color='r', alpha=0.2, marker='.')
ax.scatter(df1.iloc[:,[0,2]]['PC1'], np.full((15),+4), df1.iloc[:,[0,2]]['PC3'],color='coral',alpha=0.2, marker='.')
ax.scatter(np.full((15),-6), df1.iloc[:,[1,2]]['PC2'], df1.iloc[:,[1,2]]['PC3'],color='chocolate',alpha=0.2, marker='.')


for zdir,xs,ys,zs,ls in zip(zdir,xs,ys,zs,ls):
    ax.text(xs,ys,zs,ls,zdir, color = 'red',
             ha = 'center', va = 'center')

ax.set(xlabel = 'PC1', ylabel = 'PC2', zlabel = 'PC3')
ax.set(xlim = [-6,+3], ylim = [-4,+4], zlim = [-6,+6])
ax.view_init(elev=27)

plt.savefig('c_grafico_3D.pdf', transparent=True, orientation='landscape', pad_inches=None)
plt.show()
```
