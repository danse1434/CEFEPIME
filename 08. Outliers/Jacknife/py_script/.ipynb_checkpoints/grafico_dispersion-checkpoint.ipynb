```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages # Para varios
import numpy as np
```

```python
df1 = pd.read_csv("../Figura/pca2.csv")
df1
```

```python
fig = plt.figure(figsize=plt.figaspect(1.0))

# fig, ax = plt.subplots(nrows=2, ncols=2)

# fig = plt.figure()
gpc1 = fig.add_subplot(2,2,1)
gpc1.scatter(df1.iloc[:,[0,1,2]]['PC1'], df1.iloc[:,[0,1,2]]['PC2'])
gpc1.set(xlim = [-6,+3], ylim = [-6,+6], xlabel='PC1', ylabel='PC2')

gpc2 = fig.add_subplot(2,2,2)
gpc2.scatter(df1.iloc[:,[0,1,2]]['PC1'], df1.iloc[:,[0,1,2]]['PC3'])
gpc2.set(xlim = [-6,+3], ylim = [-6,+6], xlabel='PC1', ylabel='PC3')

gpc3 = fig.add_subplot(2,2,3)
gpc3.scatter(df1.iloc[:,[0,1,2]]['PC2'], df1.iloc[:,[0,1,2]]['PC3'])
gpc3.set(xlim = [-6,+3], ylim = [-6,+6], xlabel='PC2', ylabel='PC3')



# Plano de ejes
(xx,yy) = np.meshgrid(np.linspace(-6, 6, 10),np.linspace(-6, 6, 10))
zz = yy * 0


xs = np.array((df1.PC1))
ys = np.array((df1.PC2))
zs = np.array((df1.PC3))
zdir = np.full((15), None)
ls = np.linspace(1,14,14, dtype=np.int32)

x = np.linspace(-6, +6, 10)

ax = fig.add_subplot(2,2,4, projection='3d')
ax.plot_surface(zz,yy,xx,alpha=0.1, color = 'blue')
ax.plot_surface(xx,zz,yy,alpha=0.1, color = 'blue')
ax.plot_surface(xx,yy,zz,alpha=0.1, color = 'blue')

ax.plot(x, np.zeros(x.size), np.zeros(x.size), color = 'blue', ls='--',alpha=0.1)
ax.plot(np.zeros(x.size), x, np.zeros(x.size), color = 'blue', ls='--',alpha=0.1)
ax.plot(np.zeros(x.size), np.zeros(x.size), x, color = 'blue', ls='--',alpha=0.1)


ax.scatter(df1.iloc[:,[0,1]]['PC1'], df1.iloc[:,[0,1]]['PC2'], np.full((15),-6),color='r',alpha=0.2)
ax.scatter(df1.iloc[:,[0,2]]['PC1'], np.full((15),+4), df1.iloc[:,[0,2]]['PC3'],color='coral',alpha=0.2)
ax.scatter(np.full((15),-6), df1.iloc[:,[1,2]]['PC2'], df1.iloc[:,[1,2]]['PC3'],color='chocolate',alpha=0.2)


for zdir,xs,ys,zs,ls in zip(zdir,xs,ys,zs,ls):
    ax.text(xs,ys,zs,ls,zdir, color = 'red',
             ha = 'center', va = 'center')

#ax.scatter(df1['PC1'], df1['PC2'], df1['PC3'], marker = 'o', color = 'red')

ax.set(xlabel = 'PC1', ylabel = 'PC2', zlabel = 'PC3')
ax.set(xlim = [-6,+3], ylim = [-4,+4], zlim = [-6,+6])


#pdf = PdfPages("grafico_3D.pdf")

ax.view_init(elev=27)
#plt.savefig('grafico_3D.pdf', transparent=True)
plt.show()


#for ii in range(0,5,1):
#    ax.view_init(elev=27, azim = ii)
#    plt.show()
    #pdf.savefig( fig )
    
    #plt.savefig('grafico_%d_3D.pdf'% ii, transparent=True)
#pdf.close()

#plt.show()
```
